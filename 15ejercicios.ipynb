##ejercicio 5

```python
import math

a = float(input("Valor de a: "))
b = float(input("Valor de b: "))
c = float(input("Valor de c: "))

discriminante = b**2 - 4*a*c

x1 = (-b + math.sqrt(discriminante)) / (2*a) if discriminante >= 0 else "No tiene solución real"
x2 = (-b - math.sqrt(discriminante)) / (2*a) if discriminante >= 0 else "No tiene solución real"

print("Soluciones:", x1, x2)
```

##ejercicio 11

```python
nombre = input("Nombre: ")
peso = float(input("Peso (kg): "))
altura = float(input("Altura (m): "))

imc = peso / (altura**2)

categoria = (
    "Bajo peso" if imc < 18.5 else
    "Normal" if imc < 25 else
    "Sobrepeso" if imc < 30 else
    "Obesidad"
)

print(f"{nombre}, tu IMC es {imc:.2f} y estás en categoría: {categoria}")
```

##ejercicio 12

```python
a = float(input("Número 1: "))
b = float(input("Número 2: "))
op = input("Operación (+, -, *, /): ")

resultado = (
    a + b if op == "+" else
    a - b if op == "-" else
    a * b if op == "*" else
    a / b if op == "/" and b != 0 else
    "Error"
)

print("Resultado:", resultado)
```
