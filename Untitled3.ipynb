##IF/ELSE
Se usa cuando una condicion solo puede dar dos valores de verdad: Verdadero y Falso
```
if (condicion):
Aqui se ejecuta codigo si la condicion es verdadera

else:
Aqui se ejecuta codigo si la condicion es falsa
```

Sintaxis java
```
if (condicion){
  Aqui se ejecuta el codigo si la condicion es verdadera
}else{
  Aqui se ejecuta el codigo si la condicion es falsa
}
```




````python
edad=int(input("Cual es tu edad?"))
if edad>=18:
  print ("Usted es mayor de edad")
else:
  print("usted es menor de edad")
````

````python
if edad<10: print("Aun eres un niño")
else:
  if edad>=10 and edad<14: print("Eres un preadolescente")
  else:
    if edad>=14 and edad<18: print("eres un adolescente")
    else:
      if edad>=18 and edad<30: print("Eres un adulto joven")
      else: print("Eres un adulto")
````

##if elif
se comporta como un switch, es decir, es para multiple seleccion.

Sintaxis


```
if (C1): aqui se ejecuta codigo si C1 es TRUE
elif(C2): aqui
```



````python
if edad<10:print ("aun eres un niño")
elif edad>=10 and edad<14: print ("aun eres un preadolescente")
elif edad>=14 and edad<18: print ("eres un adolescente")
elif edad>=18 and edad<30: print ("eres un adolescente")
````
