##ejercicio 18

```python
n = int(input("Ingrese un número: "))

if n < 0:
    print("No existen factoriales negativos")
else:
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    print("El factorial es:", factorial)
```

##ejercicio 19

```python
n = int(input("Ingrese la cantidad de números Fibonacci: "))

a = 0
b = 1

for i in range(n):
    print(a)
    a, b = b, a + b
```

##ejercicio 20

```python
color_secreto = input("Usuario 1, ingrese el color secreto: ")

intentos = 0
adivinado = False

while intentos < 10 and not adivinado:
    intento = input("Adivine el color: ")
    intentos += 1

    if intento == color_secreto:
        print("¡Correcto!")
        adivinado = True
    else:
        print("Incorrecto")

if not adivinado:
    print("Se acabaron los intentos")
```

##ejercicio 23

```python
n = int(input("Ingrese un número: "))

if n <= 1:
    print("No es primo")
else:
    es_primo = True
    for i in range(2, n):
        if n % i == 0:
            es_primo = False
            break

    if es_primo:
        print("Es primo")
    else:
        print("No es primo")
```
